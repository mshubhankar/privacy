```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = None
```

```python
data=pd.read_csv('full_patients.csv')
```

```python
data.info()
```

```python
data_clean= data.iloc[:,17:]
```

```python
data_clean.drop(columns=['Molecular_Profile','Mutation','Drugs','Type'],inplace=True)
```

```python
data_clean
```

```python
data_clean.Stage=pd.Categorical(data_clean.Stage.astype('category'))
data_clean.Radio_Therapy=pd.Categorical(data_clean.Radio_Therapy.astype('category'))
data_clean.Surgery=pd.Categorical(data_clean.Surgery.astype('category'))
data_clean.Smoking=pd.Categorical(data_clean.Smoking.astype('category'))
```

```python
cat_columns = data_clean.select_dtypes(['category']).columns
```

```python
cat_columns
```

```python
data_clean[cat_columns] = data_clean[cat_columns].apply(lambda x: x.cat.codes)
```

```python
data_clean
```

```python
data_clean.to_csv('Patients_clean.csv')
```
